# **ML MODEL**
# Classification model

## **Classification Model for wine dataset**

### **Import Libraries**

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

### **Load dataset**

```python
wine = datasets.load_wine()
```

```python
wine
```

### **Input features**

```python
print(wine.feature_names)
```

### **Output features**

```python
print(wine.target_names)
```

## **Glimpse of data**

### **input features**

```python
wine.data
```

### **Output variables**

```python
wine.target
```

### **Assigning input and output variable**

```python
X = wine.data
Y = wine.target
```

```python
X.shape,Y.shape
```

## **Building classification model using random forest**

```python
wcm = RandomForestClassifier()
```

```python
wcm.fit(X,Y)
```

### **Feature importance**

```python
print(wcm.feature_importances_)
```

### **Make prediction**

```python
X[1]
```

```python
X[2]
```

```python
print(wcm.predict(X[[1]]))
```

```python
print(wcm.predict_proba(X[[1]]))
```

```python
wcm.fit(wine.data,wine.target_names[wine.target])
```

### **Data split**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
X_train.shape,Y_train.shape
```

```python
X_test.shape,Y_test.shape
```

## **Rebuild the RandomForest Model**

```python
wcm.fit(X_train,Y_train)
```

```python
print(wcm.predict(X[[2]]))
```

```python
print(wcm.predict_proba(X[[2]]))
```

### **Prediction on dataset**

```python
print(wcm.predict(X_test))
```

### **Actual class label**

```python
print(Y_test)
```

### **Model performance**

```python
print(wcm.score(X_test,Y_test))
```
