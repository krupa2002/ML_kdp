#  **ML Project**

## Load data

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
df
```

# Data Preparation

Data separation as X and Y


```python
y = df["logS"]
y
```

```python
X = df.drop('logS',axis = 1)
X
```

### data spliting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state= 100)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

# Model building

# Linear regreesion


### Training the model

```python
from sklearn.linear_model import LinearRegression
lr  = LinearRegression()
lr.fit(X_train,y_train)
```

### Applying the model to make predictiom

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_test_pred
```

### Evalulate model performance

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred)

```

```python
print(y_train.shape)
print(y_lr_train_pred.shape)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
print("LR MSE (train):",lr_train_mse)
print("LR R2 (train):",lr_train_r2)
print("LR MSE (test):",lr_test_mse)
print("LR r2 (test):",lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linearregression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns= ['Method','Training MES','Training R2','test mse','test r2']
```

```python
lr_results
```

# Random forest


### Training the model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(X_train, y_train)
```

### Applying the model to make prediction

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

### Evaluating model

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train,y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2 = r2_score(y_test,y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Linearregression',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns= ['Method','Training MES','Training R2','test mse','test r2']
```

```python
rf_results
```

# Model comparsion

```python
df_models = pd.concat([lr_results,rf_results],axis=0)
```
