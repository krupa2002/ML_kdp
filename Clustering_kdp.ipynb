### Understanding Clustering

When you're trying to learn about something, say movies, one approach might be to look for meaningful groups or collections. In machine learning too, we often group examples as a first step to understand a subject (data set) in a machine learning system. Grouping unlabeled examples is called clustering.

As the examples are unlabeled, clustering relies on unsupervised machine learning. If the examples are labeled, then clustering becomes classification.

Before you can group similar examples, you first need to find similar examples. You can measure similarity between examples by combining the examples' feature data into a metric, called a similarity measure. When each example is defined by one or two features, it's easy to measure similarity. For example, you can find similar movies by their directors. As the number of features increases, creating a similarity measure becomes more complex. For instance, you might want to find similar movies based on a combination of features like genre, director, lead actors, release year, and box office performance. The more features you consider, the more complex it becomes to determine similarity. We'll later see how to create a similarity measure in different scenarios.

# Clustering Algorithms

### Centroid-based Clustering
- **What it does**: Puts similar things together based on their average or "center".
- **Example**: If you want to group houses by location, you'd find the center of each group of houses, like the neighborhood's middle point.
- **Good for**: Quickly dividing things into groups when you know roughly where the groups are.

image.png

### Density-based Clustering
- **What it does**: Groups things based on how close they are to each other, ignoring the overall shape.
- **Example**: Imagine a crowd of people in a concert. You'd group them based on where they're densely packed, regardless of the crowd's shape.
- **Good for**: Finding areas of high activity or interest in data.

image.png

### Distribution-based Clustering
- **What it does**: Sorts things into groups based on how they fit a specific pattern, like a bell curve.
- **Example**: If you're categorizing people by height, you'd put them into groups based on how likely they are to fit into different height ranges.
- **Good for**: When your data follows a predictable pattern, like a normal distribution.

image.png

### Hierarchical Clustering
- **What it does**: Creates a family tree-like structure of groups, where groups can contain sub-groups.
- **Example**: Think of organizing a family reunion. You'd start with immediate family members, then branch out to relatives and extended family, creating a hierarchy of relationships.
- **Good for**: Organizing things in a nested or hierarchical manner.

# K-means

K-means is a popular clustering algorithm that partitions data into 'k' clusters based on feature similarity. It iteratively assigns data points to the nearest cluster center and updates the centroids until convergence. K-means is computationally efficient and effective for many clustering tasks but sensitive to initial centroid placement and outliers.

# 3 main reason why K-mean clustering

1. **Data Organization**: K-means clustering helps organize data into meaningful groups, making it easier to understand and interpret.

2. **Insight Generation**: By grouping similar data points together, K-means enables the discovery of patterns and relationships within the data, providing valuable insights.

3. **Decision Support**: Clustering with K-means aids in decision-making by segmenting data into actionable clusters, empowering businesses to make informed decisions based on identified patterns.

# Steps to implement K-mean Clustering

1. **Prepare Data**: Collect and preprocess the dataset, ensuring it is clean and formatted appropriately for clustering.

2. **Choose Number of Clusters**: Decide on the optimal number of clusters for your data, based on domain knowledge or using techniques like the elbow method.

3. **Initialize Cluster Centers**: Randomly select initial cluster centers or use a smarter initialization method.

4. **Assign Data Points to Clusters**: Calculate the distance between each data point and cluster centers, assigning each point to the nearest cluster.

5. **Update Cluster Centers**: Recalculate the centroid of each cluster based on the data points assigned to it.

6. **Repeat Until Convergence**: Iterate steps 4 and 5 until the cluster assignments stabilize and cluster centers no longer change significantly.

7. **Evaluate Clustering Results**: Assess the quality of the clustering solution using metrics like silhouette score or visual inspection.

8. **Interpret and Use Clusters**: Analyze the resulting clusters to gain insights and make data-driven decisions.

# Determining the Optimal Number of Clusters
One of the key challenges in clustering is determining the optimal number of clusters for a given dataset. This is where the Elbow Method comes in handy.

The Elbow Method:
1. Run Clustering Algorithm: Apply a clustering algorithm (e.g., k-means) to the dataset for a range of cluster numbers (k values).
2. Compute Within-Cluster Sum of Squares (WCSS): Calculate the sum of squared distances between each data point and its nearest cluster center.
3. Plot Elbow Curve: Plot the number of clusters (k) against the corresponding WCSS.
4. Identify Elbow Point: Look for the "elbow" point in the plot, where the rate of decrease in WCSS slows down significantly. This point represents the optimal number of clusters.

# **Understanding Inertia in K-Means**

Inertia, also known as within-cluster sum of squares (WCSS), is a measure used in K-means clustering to evaluate how well the data points are clustered around their respective cluster centers.

**Explanation:**

- **What is Inertia?** Inertia measures how compactly clustered the data points are around their centroids. It calculates the sum of squared distances between each data point and its assigned centroid within a cluster.

- **How it Works:** For each cluster, the distance between every data point and its centroid is calculated and squared. These squared distances are then summed up to compute the inertia for that cluster. The total inertia across all clusters is the sum of the individual cluster inertias.

- **Interpretation:** Lower inertia indicates that the data points are tightly packed around their centroids, forming cohesive clusters. Higher inertia suggests that the clusters are more spread out, indicating poorer clustering.

- **Optimization Objective:** In K-means clustering, the goal is to minimize the inertia. The algorithm iteratively adjusts the cluster centroids to minimize the sum of squared distances between data points and their assigned centroids, thereby improving the compactness of clusters.

- **Evaluation:** Inertia is used as a criterion to evaluate the quality of the clustering solution. Lower inertia values indicate better clustering performance, but it should be interpreted alongside other evaluation metrics for a comprehensive assessment of the clustering results.

### Finding the Optimal Number of Clusters

### Function Purpose

The function `find_best_clusters` determines the optimal number of clusters for K-means clustering by calculating and returning the inertia values for different numbers of clusters.

### Parameters

- `df`: The dataset to be clustered.
- `maximum_K`: The maximum number of clusters to consider.

### Steps

1. **Initialize Lists**:
   - `clusters_centers`: To store the inertia values for each number of clusters.
   - `k_values`: To store the corresponding number of clusters.

2. **Loop Through Cluster Numbers**:
   - For each number `k` from 1 to `maximum_K - 1`:
     - Create and fit a K-means model with `k` clusters.
     - Compute and store the inertia (measure of cluster compactness).
     - Store the number of clusters `k`.

3. **Return Results**:
   - Return the lists of inertia values and corresponding cluster numbers.


```python
def find_best_clusters(df, maximum_K):
    clusters_centers = []
    k_values = []
    for k in range(1, maximum_K):
        kmeans_model = KMeans(n_clusters = k)
        kmeans_model.fit(df)
        clusters_centers.append(kmeans_model.inertia_)
        k_values.append(k)
    return clusters_centers, k_values
```

```python
clusters_centers, k_values = find_best_clusters(scaled_data, 12) #finding the optimal number of clusters
```

### Elbow Method: Visualizing Optimal Clusters

```python
def generate_elbow_plot(clusters_centers, k_values): #plotting the elbow plot

    figure = plt.subplots(figsize = (12, 6)) #creating a figure and setting the size
    plt.plot(k_values, clusters_centers, 'o-', color = 'orange') #plotting the elbow plot
    plt.xlabel("Number of Clusters (K)") #naming the x-axis
    plt.ylabel("Cluster Inertia") #naming the y-axis
    plt.title("Elbow Plot of KMeans") #setting the title
    plt.show() #displaying the plot
```

```python
generate_elbow_plot(clusters_centers, k_values) #plotting the elbow plot
```
