```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/500hits/500hits.csv",encoding = "latin-1")
```

```python
df.head()
```

```python
df = df.drop(columns = ["PLAYER","CS"])
```

```python
X = df.iloc[:,0:13]
```

```python
y = df.iloc[:,13]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train ,y_test = train_test_split(X,y,random_state=17,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
kd = DecisionTreeClassifier()
```

```python
kd.get_params()
```

```python
kd.fit(X_train, y_train)
```

```python
y_pred = kd.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
kd.feature_importances_
```

```python
X.columns
```

```python
feature = pd.DataFrame(kd.feature_importances_,index = X.columns)
```

```python
feature.head(15)
```

```python
kd2 = DecisionTreeClassifier(criterion="entropy",ccp_alpha = 0.04)
```

```python
kd2.fit(X_train,y_train)
```

```python
y_pred2 = kd2.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred2))
```

```python
print(classification_report(y_test,y_pred2))
```
