**7.where(), nunique(), values_counts(), cut(), melt()**

```python
#Replace 'id' values with NaN where they are greater than or equal to 33.
import numpy as np
df['id'] = df['id'].where(df['id'] < 33, np.nan)
df
```

```python
#counting unique items 
unique_counts = df.nunique()
print(unique_counts)
```

```python
#counting values counts
value_counts=df["name"].value_counts()
value_counts
```

```python
#Bin 'id' values into specified bins using pd.cut() with bin edges [1,2,3,4,5,6].
bin_edges = [1,2,3,4,5,6]
df['id']=pd.cut(df['id'],bins= bin_edges)
df
```

```python
#Melt DataFrame, unpivoting 'target' column, creating 'total' for variable names, and 'bodyParts' for values.
import pandas as pd

melted_df = df.melt(id_vars='target', var_name='total', value_name='bodyParts')

melted_df
```

```python
import pandas as pd
import numpy as np

# Creating a DataFrame with humidity and range data
data = {'humidity': np.random.randint(40, 60, size=10),
        '2017-01-01': np.random.randint(1, 5, size=10),
        '2017-01-08': np.random.randint(1, 5, size=10)}

time1 = pd.DataFrame(data)

# Adding error bars to the bar plot
ax = time1.plot.bar(y='humidity', yerr=[time1['2017-01-01'], time1['2017-01-08']], capsize=5, rot=0)

# Setting plot labels and title
ax.set_title('Humidity with Error Bars')
ax.set_xlabel('Data Points')
ax.set_ylabel('Humidity')



```
