**4. Filtering**

```python
#filtering 
df[df['id']>4999]
```

```python
#filtering considering multiple attributes and using AND function to get all the results
df[(df['id']>4999) & (df['host_id']>9000)]
```

```python
#filtering 
df['id']>4999
```

```python
#filtering considering multiple attributes and using OR function to get all the exceptional results
df[(df['id']>4999) | (df['host_id']>9000)]
```

```python
#isin = used to filter data frames based on whether values are present in a specified Series(rows and columns)
newop=[57,157,133]
new_isin = df[df['price'].isin(newop)]
```

```python
#printing new_isin
print(new_isin)
```

```python
#negetation of isin i.e filter the exception of condition
newop=[57,157,133]
new_isin = df[~df['price'].isin(newop)]
```

```python
print(new_isin)
```

**5. Alterations,rename,reset,sort values,lowercase,contains,concat,merging**

```python
#renaming columns
df = df.rename(columns = {"minimum_nights" : "min_nights"} )
```

```python
df
```

```python
#reseting the index value based on the given condition 
filtered_df = df[df['price']>50].reset_index(drop=True)
```

```python
filtered_df
```

```python
#sorting the index values by specific attribute and in descending order 
df.sort_values(by = 'price',ascending=False)
```

```python
#making 'name' column in lowercase
df['name'].str.lower()
```

```python
#finding the string 'in' in name column
df['name'].str.contains('in')
```

```python
#finding/fetching  the string 'in' from name column 
df[df['name'].str.contains('in')]
```

```python
#concatinating the 2 attributes 'price' and 'min_nights' and adding in one dataframe based on particular condition
data_to_be_add = {"price":[45,57],"min_nights":[30,3]}
extra_info = pd.DataFrame(columns = ['price','min_nights'], data = data_to_be_add)
extra_info
```

```python
#merging 2 dataframes
df.merge(extra_info, how = 'left', on = 'price')
```

**6. Aggregations,describe,mean,value counts,group by**

```python
#describing about the dataframes (statistics)
df.describe()
```

```python
#calculating mean on 'price' column
df['price'].mean()
```

```python
#calculating value_count (total number it occured)on 'price' column
df['price'].value_counts()
```

```python
#groupby is merging two attributes and here we are calculating mean also
df.groupby('room_type')['min_nights'].mean()
```

```python
#groupby is merging two attributes and here we are counting the values also
df.groupby('room_type')['min_nights'].count()
```

**7. Plotting**

```python
#plotting the dataframe
df.plot()
```

```python
#plotting dataframe on specific attribute 'price'
df['price'].plot()
```

```python
#plotting using multiple attributes
df[['price','min_nights']].plot()
```
