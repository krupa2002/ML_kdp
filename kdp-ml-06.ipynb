# **Using pandas read_html to Web Scrape Data for Data Science**
URL to the webpage "2022-23 NBA Player Stats: Per Game

```python
year = '2023'
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
url = url_link.format(year)
url
```

Or


```python
url = 'https://www.basketball-reference.com/leagues/NBA_2023_per_game.html'
```

## **Read HTML page in pandas**

```python
import pandas as pd
```

```python
dfm= pd.read_html(url,header = 0)
```

```python
dfm
```

### Length of table

```python
len(dfm)
```

### **Selecting table 1**

```python
dfm[0]
```

```python
df2023 = dfm[0]
```

### **Data cleaning**

```python
len(df2023[df2023.Age == 'Age'])
```

```python
df = df2023.drop(df2023[df2023.Age == 'Age'].index)
```

```python
df.shape
```

### **Quick EDA**

```python
import seaborn as sns
```

Plotting histogram

```python
sns.displot(df.PTS,
            kde = False)
```

Plotting bar graph

```python
sns.distplot(df.PTS,
             kde=False,
             hist_kws=dict(edgecolor="red", linewidth=5,color = "green"))
```
